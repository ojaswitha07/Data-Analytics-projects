```python
import pandas as pd
import numpy as np
import json
```

# Load the Data

```python
df = pd.read_csv(r"C:\Users\ojaswitha\OneDrive\Documents\Desktop\CLEANING_DATA\INvideos.csv")
df.head()
```

```python
df.info()
df.describe()
```

# Check Missing Values

```python
df.isnull().sum()
```

# Handle Missing Values

```python
df['description'].fillna("No Description", inplace=True)
df['tags'].fillna("No Tags", inplace=True)
```

Missing values in textual columns were handled using appropriate placeholder values.

# Remove Duplicate Records

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

Duplicate records were removed to ensure data integrity.

# Convert Date Columns

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

Date columns were converted into datetime format for consistency.

# Remove Unnecessary Columns

```python
df.drop(columns=['thumbnail_link'], inplace=True)

```

# Handle Outliers (Views, Likes)

```python
df[['views', 'likes', 'dislikes']].describe()
```

```python
df = df[df['views'] >= 0]
df = df[df['likes'] >= 0]
```

Negative and invalid numerical values were removed to improve data quality.

# Load Category JSON File

```python
with open("IN_category_id.json") as f:
    data = json.load(f)
```

```python
categories = {}

for item in data['items']:
    categories[int(item['id'])] = item['snippet']['title']
```

# Map Category ID to Category Name

```python
df['category_name'] = df['category_id'].map(categories)
df.head()
```

# Final Cleaned Dataset

```python
df.info()
df.isnull().sum()
```

FINAL DATA CLEANING SUMMARY

Data cleaning was performed on the YouTube Trending Videos dataset to improve data quality and reliability. Missing values in textual fields were handled appropriately, duplicate records were removed, and date columns were converted to datetime format. Unnecessary columns were dropped, and category IDs were mapped to category names using JSON files. Outliers and invalid values were filtered. The cleaned dataset was saved for further analysis.
